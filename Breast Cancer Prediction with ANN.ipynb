### Model

```python
X = q.drop('label', axis = True)
y = q['label']
```

```python
from sklearn.model_selection import train_test_split

a, d, s, f = train_test_split(X, y, test_size = 0.2, random_state = 9)
```

```python
print(a.shape, s.shape, d.shape, f.shape)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

a = ss.fit_transform(a)
d = ss.transform(d)
```

```python
s
```

```python
import tensorflow as tf
tf.random.set_seed(9)
```

```python
m = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape = (30,)),
    tf.keras.layers.Dense(50, activation = 'relu'),
    tf.keras.layers.Dense(2, activation = 'sigmoid')
])

m.compile(optimizer = 'adam',
         loss = 'sparse_categorical_crossentropy',
         metrics = ['accuracy'])
```

```python
h = m.fit(a,s, validation_split = 0.1, epochs = 50)
```

```python
plt.figure(figsize =(9,6))
plt.plot(h.history['accuracy'], c = 'r', label = 'accuracy')
plt.plot(h.history['val_accuracy'], c = 'g', label = 'val_accuracy')
plt.legend()

plt.figure(figsize =(9,6))
plt.plot(h.history['loss'], c = 'y', label = 'loss')
plt.plot(h.history['val_loss'], c = 'b', label = 'val_loss')
plt.legend();
```

```python
g = m.predict(d)
```

```python
import numpy as np

h = [np.argmax(i) for i in g]
len(h)
```

```python
from sklearn.metrics import accuracy_score

print('accuracy_score : ',accuracy_score(f,h))
```
